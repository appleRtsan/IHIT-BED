```python
import pandas as pd
import matplotlib.pyplot as plt
from pytorch_tabnet.tab_model import TabNetClassifier
from sklearn.metrics import roc_auc_score

```

```python
loaded_clf = TabNetClassifier()
loaded_clf.load_model("BC classifier/model.zip")
```

```python
# feature indexing for BC classifier
fi = []
with open ("BC classifier/feature indexing.txt",'r') as f:
    
    while True:
        s = f.readline().strip('\n')
        if not s:
            break
        # print(s)
        fi.append(str(s))
```

```python
df_test = pd.read_excel('example cohort.xlsx') # second independent testing cohort


for i in range(len(df_test)):
    if df_test.loc[i,'CRP'] > 10.0 :
        df_test.loc[i,'CRP_class'] = 1
    else:
        df_test.loc[i,'CRP_class'] = 0
        
X_test = df_test[fi].values
y_test = df_test['Blood Culture'].values
```

```python
preds = loaded_clf.predict(X_test)

print(round(roc_auc_score(df_test['Blood Culture'].values, preds),2))
```

```python
loaded_clf_CRP = TabNetClassifier()
loaded_clf_CRP.load_model("CRP classifier/model.zip")
CRP_features = [ col for col in df_test.columns if col not in ["CRP_class"]+["CRP"]+['Barcode']+['Blood Culture']+['Blood Culture_R']+['preds_b10']+['preds_b1']+['preds']]
X_test = df_test[CRP_features].values
y_test = df_test['CRP_class'].values
```

```python
preds_CRP = loaded_clf_CRP.predict_proba(X_test)[:,-1]

print(round(roc_auc_score(df_test['CRP_class'].values, preds_CRP),2))
```
